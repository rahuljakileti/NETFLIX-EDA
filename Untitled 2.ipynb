```python
# Load and explore Netflix data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('netflix_titles.csv')
print('Dataset loaded successfully')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic data info
print('Data types:')
print(df.dtypes)
print('\
Missing values:')
print(df.isnull().sum())
print('\
Basic statistics:')
df.describe(include='all')
```

```python
# Data cleaning
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)

# Content type distribution
plt.figure(figsize=(10, 6))
type_counts = df['type'].value_counts()
plt.subplot(1, 2, 1)
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%', colors=['#E50914', '#221F1F'])
plt.title('Content Type Distribution')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='type', palette=['#E50914', '#221F1F'])
plt.title('Movies vs TV Shows')
plt.tight_layout()
plt.show()
```

```python
# Release year trends
plt.figure(figsize=(12, 6))
release_year_counts = df['release_year'].value_counts().sort_index()
plt.plot(release_year_counts.index, release_year_counts.values, color='#E50914', linewidth=2)
plt.title('Netflix Content by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True, alpha=0.3)
plt.show()

# Content added over time
plt.figure(figsize=(12, 6))
yearly_additions = df.groupby('year_added').size()
plt.bar(yearly_additions.index, yearly_additions.values, color='#E50914', alpha=0.7)
plt.title('Content Added to Netflix by Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Top countries
plt.figure(figsize=(12, 8))
top_countries = df['country'].value_counts().head(15)
plt.barh(range(len(top_countries)), top_countries.values, color='#E50914')
plt.yticks(range(len(top_countries)), top_countries.index)
plt.title('Top 15 Countries by Content Count')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()

# Rating distribution
plt.figure(figsize=(12, 6))
rating_counts = df['rating'].value_counts()
plt.bar(rating_counts.index, rating_counts.values, color='#E50914', alpha=0.7)
plt.title('Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Duration analysis for movies vs TV shows
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(movies['duration_num'].dropna(), bins=30, color='#E50914', alpha=0.7, edgecolor='black')
plt.title('Movie Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(tv_shows['duration_num'].dropna(), bins=15, color='#221F1F', alpha=0.7, edgecolor='black')
plt.title('TV Show Seasons Distribution')
plt.xlabel('Number of Seasons')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
duration_stats = df.groupby('type')['duration_num'].describe()
print('Duration Statistics by Type:')
print(duration_stats)

plt.tight_layout()
plt.show()
```

```python
# Genre analysis
genres_split = df['listed_in'].str.split(', ').explode()
top_genres = genres_split.value_counts().head(15)

plt.figure(figsize=(12, 8))
plt.barh(range(len(top_genres)), top_genres.values, color='#E50914')
plt.yticks(range(len(top_genres)), top_genres.index)
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()

# Correlation heatmap for numerical variables
plt.figure(figsize=(10, 8))
numerical_cols = ['release_year', 'year_added', 'month_added', 'duration_num']
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', center=0, square=True)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Cast analysis - most appeared actors
cast_split = df['cast'].dropna().str.split(', ').explode()
top_actors = cast_split.value_counts().head(15)

plt.figure(figsize=(12, 8))
plt.barh(range(len(top_actors)), top_actors.values, color='#E50914')
plt.yticks(range(len(top_actors)), top_actors.index)
plt.title('Top 15 Most Appeared Cast Members on Netflix')
plt.xlabel('Number of Appearances')
plt.gca().invert_yaxis()
plt.show()

print('Top 10 Most Appeared Cast Members:')
for i, (actor, count) in enumerate(top_actors.head(10).items(), 1):
    print(f'{i}. {actor}: {count} appearances')
```
